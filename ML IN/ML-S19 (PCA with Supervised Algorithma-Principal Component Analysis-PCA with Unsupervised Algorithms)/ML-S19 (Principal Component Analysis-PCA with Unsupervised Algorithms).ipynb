## Scaling Data

```python
from sklearn.preprocessing import StandardScaler

# PCA'de satandard scale olmaz ise olmazdır. Tüm featurların std'sı 1 yapılarak varianceları (std'nın karesi) da 1 yapılıyor. 
# Tüm featurların varience (yani ne kadar bilgi taşıdıkları) 1'e sabitlenerek PCA dönüşüm algortimasının tüm featurlara eşit 
# davranmasını sağlıyoruz. Varienceları 1 olmaz ise PCA varience büyük olan featurlardan daha fazla katkı alma eğiliminde 
# olacaktır ve doğal olarak elde edeceğimiz componenetler yanlış olacaktır.
```

```python
scaler = StandardScaler()
```

```python
scaled_X = scaler.fit_transform(X)
```

```python
X_scaled_df = pd.DataFrame(scaled_X, columns = X.columns)
X_scaled_df.head()
```

```python
plt.figure(figsize = (25, 9))
sns.heatmap(X_scaled_df.corr().round(2), annot = True);

# PCA algortiması featurelar arasında anlamlı correlasyonlar varsa çok daha iyi çalışacağından corr'ları kontrol ediyoruz.
# PCA aynı zamanda multicollineraty sorununu da halleder.
# Biribirleri ile yüksek corr. veya anlamlı corr ilişkisi olan featurlar daha yoğun bir şekilde aynı component içerisinde
# birleştirilirler.
# PCA'de elde edeceğimiz tüm yeni componetlerde az ya da çok mutlaka tüm featurların katkısı olur.
```

## PCA

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)

# n_component default olarak None'dır. datamızdaki feature veya satır sayısından hangisi az ise PCA maksimum o sayı kadar
# component oluşturur.

# PCA'in çalışma mantığını anlamamız bakımından 2 component tanımlıyoruz.
```

```python
principal_components = pca.fit_transform(scaled_X)

# fit işlemi ile datamızda ki feature/row sayısından hangisi az ise o sayı kadar her bir component için eigen vektörler 
# belirlenir. Transform işlemi ile eigen vektörler ve her bir gözleme ait feature değerleri karşılıklı çarpılıp toplanarak 
# componenetler elde edilir.
# n_componenets olarak belirttiğimiz sayı kadar component döndürülür.
```

```python
principal_components

# n_components=2 belirttiğimizden sadece 2 component döndürüldü.
```

```python
component_X = pd.DataFrame(data = principal_components, columns = ["first_component", 
                                                                    "second_component"])
component_X

# daha okunaklı hale gelmesi için componentlerimizi isimlendirdik ve component_X değişkenine atadık.
```

## K-Means Algorithm

```python
from sklearn.neighbors import BallTree
import numpy as np
import pandas as pd


def hopkins(data_frame, sampling_size):
    """Assess the clusterability of a dataset. A score between 0 and 1, a score around 0.5 express
    no clusterability and a score tending to 0 express a high cluster tendency.

    Parameters
    ----------
    data_frame : numpy array
        The input dataset
    sampling_size : int
        The sampling size which is used to evaluate the number of DataFrame.

    Returns
    ---------------------
    score : float
        The hopkins score of the dataset (between 0 and 1)

    Examples
    --------
    >>> from sklearn import datasets
    >>> from pyclustertend import hopkins
    >>> X = datasets.load_iris().data
    >>> hopkins(X,150)
    0.16
    """

    if type(data_frame) == np.ndarray:
        data_frame = pd.DataFrame(data_frame)

    # Sample n observations from D : P

    if sampling_size > data_frame.shape[0]:
        raise Exception(
            'The number of sample of sample is bigger than the shape of D')

    data_frame_sample = data_frame.sample(n=sampling_size)

    # Get the distance to their neirest neighbors in D : X

    tree = BallTree(data_frame, leaf_size=2)
    dist, _ = tree.query(data_frame_sample, k=2)
    data_frame_sample_distances_to_nearest_neighbours = dist[:, 1]

    # Randomly simulate n points with the same variation as in D : Q.

    max_data_frame = data_frame.max()
    min_data_frame = data_frame.min()

    uniformly_selected_values_0 = np.random.uniform(min_data_frame[0], max_data_frame[0], sampling_size)
    uniformly_selected_values_1 = np.random.uniform(min_data_frame[1], max_data_frame[1], sampling_size)

    uniformly_selected_observations = np.column_stack((uniformly_selected_values_0, uniformly_selected_values_1))
    if len(max_data_frame) >= 2:
        for i in range(2, len(max_data_frame)):
            uniformly_selected_values_i = np.random.uniform(min_data_frame[i], max_data_frame[i], sampling_size)
            to_stack = (uniformly_selected_observations, uniformly_selected_values_i)
            uniformly_selected_observations = np.column_stack(to_stack)

    uniformly_selected_observations_df = pd.DataFrame(uniformly_selected_observations)

    # Get the distance to their neirest neighbors in D : Y

    tree = BallTree(data_frame, leaf_size=2)
    dist, _ = tree.query(uniformly_selected_observations_df, k=1)
    uniformly_df_distances_to_nearest_neighbours = dist

    # return the hopkins score

    x = sum(data_frame_sample_distances_to_nearest_neighbours)
    y = sum(uniformly_df_distances_to_nearest_neighbours)

    if x + y == 0:
        raise Exception('The denominator of the hopkins statistics is null')

    return x / (x + y)[0]
```

```python
from sklearn.cluster import KMeans

ssd = []

K = range(2,10)

for k in K:
    model = KMeans(n_clusters =k, random_state=42)
    model.fit(component_X)
    ssd.append(model.inertia_)
```

```python
plt.plot(K, ssd, "bo-")
plt.xlabel("Different k values")
plt.ylabel("inertia-error") 
plt.title("elbow method")
```

```python
ssd
```

```python
pd.Series(ssd).diff()

# keskin düşüşün durduğu ilk noktanın 3 olduğunu görüyoruz. 
```

```python
from yellowbrick.cluster import KElbowVisualizer

model_ = KMeans(random_state=42)
visualizer = KElbowVisualizer(model_, k=(2,9))

visualizer.fit(component_X)        # Fit the data to the visualizer
visualizer.show();

# yellowbrick elbow method anlamlı düşüşleri de dikkate alarak kırılmanın yataylaşmaya başladığı noktayı seçme eğilimindedir.
# Normalde 5 noktasına kadar anlamı düşüşlerin olduğunu görüyoruz ama burda 4 noktasını seçmiş. Bunun sebebi k-means seçilecek
# olan k noktasını belirlerken silhoutte scorunu da dikkate alır. Her nekadr anlamlı düşüş 5 noktasında duruyor olsa da
# aşağıda gördüğünüz gibi 5 noktasındaki silhoutte scoru çok düşük olduğundan 5 noktasını değil 4 noktasını seçmiş.
```

```python
from sklearn.metrics import silhouette_score

range_n_clusters = range(2,9)
for num_clusters in range_n_clusters:
    # intialise kmeans
    kmeans = KMeans(n_clusters=num_clusters, random_state=42)
    kmeans.fit(component_X)
    cluster_labels = kmeans.labels_
    # silhouette score
    silhouette_avg = silhouette_score(component_X, cluster_labels)
    print(f"For n_clusters={num_clusters}, the silhouette score is {silhouette_avg}")
```

```python
sns.scatterplot(x = component_X.first_component, y= component_X.second_component, alpha=0.7);

# aşağıdaki componentlerimizin görseline bakacak olursak her ne kdar 2 küme için silhouute scoru yüksek olsa da inertia değeri
# çok yüksek ve kümelerden bir tanesi çok geniş bir alana yayılmış olacak.

# 3 veya 4 küme sayıları mantıklı görünüyor. 2'sini de seçebiliriz.

# Classic Elbow metodu, yellowbrick elbow metodu, silhoutte scoru ve görselleştirme sonucunda 3 veya 4 kümenin de olabileceğine 
# karar verdik ve son karar için müşteri ile görüşüp nihai kararı vereceğiz. Müşteri ile görüşünce müşteri datayı uzmana 
# inceletip bu datada kanser kanser değil şeklinde 2 grup olduğunu söyledi. Bu durumda yaptığımız tüm çalışmaları rafa
# kaldırıp 2 cluster ile devam ediyoruz.
```

```python
model = KMeans(n_clusters =2, random_state=42)
clusters = model.fit_predict(component_X)
clusters

# uzman desteği ışığında 2 cluster ile devam modelimizi kuruyoruz.
```
